```python
output = pd.read_csv('output_new.csv')
vehicle_master = pd.read_csv('vehicle_master.csv')
#add new dummy columns
output['Size'] = 0
output['p_year'] = 0
output['Ftype'] = 0
output['ratio'] = 0
output['Demand'] = 0
output['Demand_id'] = 0
output['age'] = 0
output['k'] = 0
output['T_val'] = 0
```

```python
output
```

```python
#delete rows with 0 in solver val
output = output[output['Solver_Value'] >= 0]
output.reset_index(drop=True, inplace=True)
output
```

```python
for i in range(len(output['Solver_Var'])):
    if output['Solver_Var'][i][0] == 'T':
        output['T_val'][i] = 'U' + output['Solver_Var'][i][1:]
```

```python
t_df = output[output['T_val'] != 0] #t_df is the dataframe with all the T values
t_df.reset_index(drop=True, inplace=True)
t_df
```

```python
t_df.to_csv('t_df.csv', index=False)
```

```python
#reomve all T in output
output = output[output['T_val'] == 0]
output.reset_index(drop=True, inplace=True)
output
```

```python
#convert first column to int
output['Solver_Var'] = output['Solver_Var'].astype(str)
```

```python
output
```

```python
for i in range(len(output['Solver_Var'])):
    #Year columns
    if output['Solver_Var'][i][3] == ',':
        output['Year'][i] = int(output['Solver_Var'][i][2])+2023
        x = 5
    else:
        output['Year'][i] = int(output['Solver_Var'][i][2:4])+2023
        x = 6
    
    if output['Solver_Var'][i][0] == 'P':
        output['p_year'][i] = output['Year'][i]
    else:
        if output['Solver_Var'][i][x] == ',':
            output['p_year'][i] = int(output['Solver_Var'][i][x-1])+2023
        else:
            output['p_year'][i] = int(output['Solver_Var'][i][x-1:x+1])+2023   

    #Type columns and NUM_VEHICLES
    if output['Solver_Var'][i][0] == 'P':
        output['Type'][i] = 'Buy'
        output['Num_Vehicles'][i] = math.ceil(output['Solver_Value'][i])
        j = -2
    elif output['Solver_Var'][i][0] == 'S':
        output['Type'][i] = 'Sell'
        output['Num_Vehicles'][i] = math.ceil(output['Solver_Value'][i]) #max(1,math.floor(output['Solver_Value'][i]))
        j = -2
    elif output['Solver_Var'][i][0] == 'U':
        output['Type'][i] = 'Use'
        output['Num_Vehicles'][i] = (t_df.loc[t_df['T_val'] == output['Solver_Var'][i], 'Solver_Value'].values[0]).astype(int)
        j = -4 
    else:
        output['Type'][i] = 'Fleet'
        output['Num_Vehicles'][i] = math.ceil(output['Solver_Value'][i])
        j = -2

    #Fuel
    if int(output['Solver_Var'][i][j]) == 1 :
        output['Fuel'][i] = 'LNG'
        output['Ftype'][i] = 'LNG'
        output['k'][i] = 1

    elif int(output['Solver_Var'][i][j]) == 2 :
        output['Fuel'][i] = 'BioLNG'
        output['Ftype'][i] = 'LNG'
        output['k'][i] = 2

    elif int(output['Solver_Var'][i][j]) == 3 :
        output['Fuel'][i] = 'HVO'
        output['Ftype'][i] = 'Diesel'
        output['k'][i] = 3

    elif int(output['Solver_Var'][i][j]) == 4 :
        output['Fuel'][i] = 'B20'
        output['Ftype'][i] = 'Diesel'
        output['k'][i] = 4
 
    elif int(output['Solver_Var'][i][j]) == 5 :
        output['Fuel'][i] = 'Electricity'
        output['Ftype'][i] = 'BEV'
        output['k'][i] = 5

    elif int(output['Solver_Var'][i][j]) == 6 :
        output['Fuel'][i] = 'Electricity'
        output['Ftype'][i] = 'BEV'
        output['k'][i] = 6

    elif int(output['Solver_Var'][i][j]) == 7 :
        output['Fuel'][i] = 'Electricity'
        output['Ftype'][i] = 'BEV'
        output['k'][i] = 7

    else:
        output['Fuel'][i] = 'Electricity'
        output['Ftype'][i] = 'BEV'
        output['k'][i] = 8

    #Update distance bucket
    if output['Solver_Var'][i][0] == 'U':
        if int(output['Solver_Var'][i][-2]) == 0 :
            output['Distance_bucket'][i] = 'D1'
        elif int(output['Solver_Var'][i][-2]) == 1 :
            output['Distance_bucket'][i] = 'D2'
        elif int(output['Solver_Var'][i][-2]) == 2 :
            output['Distance_bucket'][i] = 'D3'
        else:
            output['Distance_bucket'][i] = 'D4'
  

    #Extra shape and Ftype
    if int(output['Solver_Var'][i][j-2]) == 0 :
        output['Size'][i] = 'S1'
    elif int(output['Solver_Var'][i][j-2]) == 1 :
        output['Size'][i] = 'S2'
    elif int(output['Solver_Var'][i][j-2]) == 2 :
        output['Size'][i] = 'S3'
    else:
        output['Size'][i] = 'S4'

    #Creating ID column
    output['ID'][i] = output['Ftype'][i] + '_' + output['Size'][i] + '_' + str(output['p_year'][i])
    output['ratio'][i] =  min(output['Solver_Value'][i] / output['Num_Vehicles'][i],1)
    #output['Num_Vehicles'][i] = math.ceil(output['Solver_Value'][i]) # #

    output['Demand_id'][i] = output['Size'][i] + '_' + str(output['Distance_bucket'][i]) + '_' + str(output['Year'][i])

    #Distance per vehicle
    if output['Type'][i] == 'Use':
        output['Distance_per_vehicle(km)'][i] = (vehicle_master.loc[vehicle_master['ID'] == output['ID'][i], 'Yearly range (km)'].values[0] * output['ratio'][i])
        # if output['Demand_id'][i] == 'S4_D1_2036' : 
        #     print("rascal")
        #     output['Distance_per_vehicle(km)'][i] = output['Distance_per_vehicle(km)'][i] * 1.00005

    
    output['Demand'][i] = output['Num_Vehicles'][i] * output['Distance_per_vehicle(km)'][i]
    output['age'][i] = output['Year'][i] - output['p_year'][i]+1
```
